```python
# there are 5 subjects -psychology,nano technology,medicine,oncology,anthropology . 2 students have applied to do masters in any of these 5 fields with the line plot
#show in which they can excel. the cutoff mrks for each subject is 70

 #student 1-[67,45,73,71,57]
#student 2-[43,67,68,53,85]

sub=['psycho','nano_tech','medic','onco','anthro']
std1=[67,45,73,71,57]
std2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]

#student 1
#plt.plottype(X,Y,label,color,marker,linewidth,linestyle)
plt.plot(sub,std1,label='student 1',color='k',marker='P')  #plot-->line, #scatter---->dots

# student 2
plt.plot(sub,std2,label='student 2',color='b',marker='H')

#cutoff
plt.plot(sub,cutoff,label='cutoff',color='r',linestyle=':',linewidth=4)
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
# An ind VS Aus match is on live,india has given 298 as target for aus the runs for every 5 overs is given as follows
#ind runs=25,51,84,131,160,189,220,250,267,297
#aus have completed only upto 25 overs
#aus runs=15,41,94,110,151
#plot overs Vs runs for both the countries with ind as blue line and aus as yellow line with required labels and legend.The marker for ind is 'flag'.The marker for aus is 'plus'
#take overs from arange function and create runs as arrays

from matplotlib import pyplot as plt
import numpy as np
ind_runs=np.array([25,51,84,131,160,189,220,250,267,297])
aus_runs=np.array([15,41,94,110,151])
ind_overs=np.arange(5,51,5)
aus_overs=np.arange(5,26,5)

#india
plt.plot(ind_overs,ind_runs,label='india',color='blue',marker='>')

#aus
plt.plot(aus_overs,aus_runs,label='austrialia',color='yellow',marker='+')

plt.title("IND VS AUS")
plt.grid()
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.legend()
print(ind_overs)
```
