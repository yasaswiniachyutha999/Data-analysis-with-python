```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
tb="https://en.wikipedia.org/wiki/List_of_tallest_buildings_in_India"
```

```python
page=requests.get(tb)
```

```python
code=BeautifulSoup(page.text,'html')
print(code)
```

```python
table=code.find('table')
print(table)
```

```python
h=table.find_all('th')
```

```python
headings=[i.text for i in h]
print(headings)
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(columns=headings)
print(df1)
```

```python
rows=table.find_all('tr')
print(rows)
```

```python
data=table.find_all('td')
```

```python
data_act=[i.text for i in data]
print(data_act)
```

```python
st=0
sp=9
for i in range(len(data_act)//9):
  df1.loc[i]=data_act[st:sp]
  st=st+9
  sp=sp+9
df1
```

```python
df1.to_csv("listofbul.csv")
```
