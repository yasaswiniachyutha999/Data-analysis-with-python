**Extracting html code of a webpage**

program flow
Import libraries
save the urf
Using requests.get(),access the web page
Using BeautifulSoup(),access the HTML code


```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
gc="https://www.w3schools.com/"
```

```python
page=requests.get(gc)
```

```python
code=BeautifulSoup(page.text,'html')
print(code)
```

```python
tit=code.find('title')
print(tit)
```

```python
print(tit.text)
```

**Retriving information of top 10 richest people from forbes**

```python
forbes="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1#:~:text=Mr%20Mukesh%20Ambani%2C%20the%20Managing,lakh%20crore%20(%24104%20billion)."
```

```python
p=requests.get(forbes)
```

```python
soup=BeautifulSoup(p.text,'html')
print(soup)
```

```python
tab=soup.find('table')
print(tab)
```

```python
h=tab.find_all('th')
print(h)
```

```python
heading=[i.text for i in h]
print(heading)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=heading)
print(mydf)
```

```python
rows=tab.find_all('tr')
print(rows)
```

```python
data=tab.find_all('td')
print(data)
```

```python
data_act=[i.text for i in data]
print(data_act)
```

```python
data_act[0:4]
```

```python
st=0
sp=4
for i in range(len(data_act)//4):
  mydf.loc[i]=data_act[st:sp]
  st=st+4
  sp=sp+4
mydf
```

```python
mydf.to_csv("tabledata.csv")
```
