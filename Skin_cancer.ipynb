```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#define image size and batch size
IMG_SIZE=224
BATCH_SIZE=32
```

```python
#creating traning data parameters
train_datagen=ImageDataGenerator(rescale=1./255,validation_split=0.2)
```

```python
#creating training data with above parameters
#folder =parameters.flow_from_directory(path,ts,bs,cm,subset)
train_generator=train_datagen.flow_from_directory("/content/drive/MyDrive/train",
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)
```

```python
#creating validation data
val_generator=train_datagen.flow_from_directory("/content/drive/MyDrive/train",
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
```

```python
#define model
model=keras.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(IMG_SIZE,IMG_SIZE,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D (64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D (128,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(1,activation='sigmoid'),
])
```

```python
# compile the model

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_generator,validation_data=val_generator,epochs=8)
```

```python
model.save("model.h5","label.txt")
```

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
# load the saved model
model = load_model("/content/model.h5")
```

```python
# load and preprocess the test image

test_image_path =("/content/drive/MyDrive/train/benign/100.jpg")
img=image.load_img(test_image_path, target_size=(224,224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array,axis=0)
```

```python
# add batch dimension
img_array /=225.        # normalize the pixel values
# make predictions
prediction = model.predict(img_array)
# print the prediction
print(prediction)
```

```python
if prediction < 0.5:
  print("Prediction : No skin cancer (Probability:",prediction[0][0],")")
else:
  print("Prediction : skin cancer present (Probability:",prediction[0][0],")")
```
