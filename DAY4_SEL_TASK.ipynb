```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
options=webdriver.ChromeOptions()
```

```python
driver = webdriver.Chrome(options=options)
```

```python
driver.get('https://www.amazon.in/')
```

```python
search = driver.find_element(By.XPATH,".//input[@class='nav-input nav-progressive-attribute']")
```

```python
search.send_keys("Dell Laptops")
```

```python
search.send_keys(Keys.ENTER)
```

```python
company = driver.find_elements(By.XPATH,".//span[@class='a-size-medium a-color-base a-text-normal']")#find elements give in tags
companies=[i.text for i in company]
print(companies)
print(len(companies))
```

```python
price = driver.find_elements(By.XPATH,".//span[@class='a-price-whole']")
prices=[i.text for i in price]
print(prices)
print(len(prices))
```

```python
star = driver.find_elements(By.XPATH,".//i[@class='a-icon a-icon-prime a-icon-medium']")
stars=[i.text for i in star]
print(stars)
print(len(stars))
```

```python
df1=pd.DataFrame(columns=["Company","Price","Desc"])
df1
```

```python

df1['Company']= companies[0:10]
df1['Price'] = prices[0:10]
df1['Stars'] = stars[0:10]
```

```python
df1
```

```python
df1.to_csv("mydf1.csv")
```

```python
options=webdriver.ChromeOptions()
#options.add_argument('--some-option')
#Create WebDriver instance with options
driver = webdriver.Chrome(options=options)
driver.get('https://www.myntra.com/')
search = driver.find_element(By.XPATH,".//input[@class='desktop-searchBar']")
search.send_keys("smart watches")
search.send_keys(Keys.ENTER)
```

```python
brand = driver.find_elements(By.XPATH,".//h3[@class='product-brand']")#find elements give in tags
brands=[i.text for i in brand]#without tags
print(brands)
print(len(brands))
```

```python
desc = driver.find_elements(By.XPATH,".//h4[@class='product-product']")
descs=[i.text for i in desc]
print(descs)
print(len(descs))
```

```python
price = driver.find_elements(By.XPATH,".//span[@class='product-discountedPrice']")
prices=[i.text for i in price]
print(prices)
print(len(prices))
```

```python
df=pd.DataFrame(columns=["Brand","Desc","Price"])#created dataframe with cols
df
```

```python
df['Brand']= brands
df['Desc'] = descs
df['Price'] = prices
df
```

```python
df.to_csv('mydf2.csv')
```
